```python
#gender distribution
#count plot for automatic counting the frequency
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()

```

```python
#from the above chart we have analysed that:
#the number of females in the data is more then the number of males
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
#analyze that there is any effect on the student performace
# based on parent eductaion
plt.figure(figsize=(5,5))
sns.heatmap(data=gb,annot=True,linewidth='0.05')
plt.show()
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',
                                            "ReadingScore":'mean',
                                            "WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(data=gb1,annot=True,linewidth='0.05')
plt.title("general distributation")
plt.show()
```

```python
#from the above chart we have concluded that there is no negative impact
#on the student's performance based on the parent marital status
```

```python
#distribution of ethnicGroup(percentage Distribution)
groupA=df.loc[(df['EthnicGroup'] == 'Group A')].count()
#new data frame are created that containe the only  values come under group A

groupB=df.loc[(df['EthnicGroup'] == 'Group B')].count()
ethnic_percentages = (groupA / groupA.sum()) * 100

groupC=df.loc[(df['EthnicGroup'] == 'Group C')].count()
groupD=df.loc[(df['EthnicGroup'] == 'Group D')].count()
groupE=df.loc[(df['EthnicGroup'] == 'Group E')].count()

mlist=["groupA","groupB","groupC","groupD","groupE"]
label=["Group A","Group B","Group C","Group D","Group E"]
#mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(ethnic_percentages,labels=label)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
